```python
# @title Default title text
import sqlite3
import pandas as pd
import io
from IPython.display import display
import ipywidgets as widgets

# מילון לאחסון מסדי נתונים
databases = {}

# אזור להצגת תוצאות
output = widgets.Output()

# פונקציה להעלאת קובץ CSV ויצירת מסד נתונים
def upload_csv(change):
    uploaded_file = change['new']
    if uploaded_file:
        try:
            # קריאת קובץ CSV ל-DataFrame
            file_name = list(uploaded_file.keys())[0]
            file_content = uploaded_file[file_name]['content']
            df = pd.read_csv(io.BytesIO(file_content))

            # יצירת מסד נתונים חדש
            db_name = file_name.split('.')[0] + ".db"
            db_path = f"/content/{db_name}"
            conn = sqlite3.connect(db_path)
            df.to_sql("data", conn, if_exists="replace", index=False)
            conn.close()

            # הוספת מסד הנתונים למילון
            databases[file_name] = db_path
            db_selector.options = databases.keys()

            with output:
                print(f"Database '{file_name}' created successfully!")
        except Exception as e:
            with output:
                print("Error:", e)

# רכיב העלאת קבצים
uploader = widgets.FileUpload(
    accept=".csv",
    multiple=False
)
uploader.observe(upload_csv, names='value')

# פונקציה להרצת שאילתות
def run_query(db_name, query):
    try:
        db_path = databases[db_name]
        conn = sqlite3.connect(db_path)
        result = pd.read_sql(query, conn)
        output.clear_output()
        with output:
            display(result)
        conn.close()
    except Exception as e:
        output.clear_output()
        with output:
            print("Error:", e)

# תפריט בחירת מסד נתונים
db_selector = widgets.Dropdown(
    options=databases.keys(),
    description="Database:",
    layout=widgets.Layout(width="50%")
)

# רכיב להזנת שאילתה רב-שורתית
query_input = widgets.Textarea(
    value="SELECT * FROM data",
    description="SQL Query:",
    layout=widgets.Layout(width="80%", height="100px"),
)

# כפתור להרצת השאילתה
run_button = widgets.Button(description="Run Query")

# פעולה לכפתור
def on_run_button_clicked(b):
    db_name = db_selector.value
    query = query_input.value
    run_query(db_name, query)

run_button.on_click(on_run_button_clicked)

# הצגת הממשק
display(widgets.VBox([uploader, db_selector, query_input, run_button, output]))
```
